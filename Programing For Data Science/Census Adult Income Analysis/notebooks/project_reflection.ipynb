# Project Reflection

## 1. General Conclusion

To achieve high income (>$50K) in the 1994 US economy, the "success formula" relied on four key pillars:

- **Education is Key:** A Bachelor's degree served as the decisive tipping point for income breakthroughs.

- **Stability as a Catalyst:** Marriage was a robust financial indicator, predicting wealth even better than age alone.

- **Capital over Labor:** Investment income outweighed excessive manual labor. Working extreme hours (>60h/week) showed diminishing returns.

- **Skills over Origin:** Expertise mattered more than nationality; high-skilled immigrant groups often outperformed US natives.

**Final Summary:** The 1994 Census dataset provides more than just statistics; it offers a vivid snapshot of the American social landscape. It demonstrates that achieving the >$50,000 dream in the mid-90s was best secured through a combination of advanced education (Human Capital), social stability (Marriage), and strategic asset management (Capital Gains), rather than relying solely on increasing manual labor hours.

## 2. Challenges & Difficulties
### Lê Nguyên Thảo:
- **The Obstacle**: The most significant challenge was identifying the analytical blind spots inherent in each research question. I realized that every initial hypothesis contains hidden biases that can severely skew the final prediction

- **Overcoming it**: To address this, I had to develop specific robustness checks for each question to ensure objectivity. I learned to dig deeper to ensure our conclusions were grounded in reality rather than statistical artifacts.
### Nguyễn Thanh Owen:
- **The Obstacle**: The main obstacle was handling imbalanced data, especially for underrepresented groups such as the Amer-Indian-Eskimo population. Another challenge was correctly understanding and applying survey weights to adjust raw data.

- **Overcoming it**: These issues were addressed by using weighted analysis instead of raw counts, ensuring that percentage estimates better reflected the true population distribution. Careful comparison between weighted and unweighted results was also conducted to confirm that observed discrepancies were due to sampling bias rather than calculation errors.

### Trần Minh Quang:
- **The Obstacle**: 
    * **High Cardinality & Sparsity:** The native-country variable contained 40+ distinct values with sparse distribution, creating statistical noise.
    * **Continuous Data Noise:** The hours-per-week variable was continuous and exhibited a non-linear relationship with income, making direct correlation difficult to interpret.
- **Overcoming it**:
    * **Regional Grouping:** I aggregated countries into broader Geographic/Economic Regions (e.g., Europe, Asia). This reduced dimensionality and highlighted the "Immigrant Advantage."
    * **Logical Binning:** I discretized work hours into functional categories (Part-time, Full-time, Overtime, Excessive). This revealed that working "Excessive" hours (>60h) often yields diminishing returns compared to moderate "Overtime."
